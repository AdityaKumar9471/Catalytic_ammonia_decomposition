```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['Catalyst'] = le.fit_transform(df['Catalyst'])

df["Supported by"] = le.fit_transform(df["Supported by"])

```

```python
X=df.drop(columns="Ammonia conversion (%)")
y=df["Ammonia conversion (%)"]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
def evaluate_model(model):
  y_pred=model.predict(X_test)
  from sklearn.metrics import r2_score, mean_squared_error,mean_absolute_error
  R2=r2_score(y_test, y_pred)
  RMSE=mean_squared_error(y_test, y_pred, squared=False)
  MAE=mean_absolute_error(y_test, y_pred)
  print("Result for testing sample")
  print(f"R2 score: {R2}")
  print(f"MSE: {RMSE}")
  print(f"MAE: {MAE}")

def evaluate_model_train(model):
  y_pred=model.predict(X_train)
  from sklearn.metrics import r2_score, mean_squared_error,mean_absolute_error
  R2=r2_score(y_train, y_pred)
  RMSE=mean_squared_error(y_train, y_pred, squared=False)
  MAE=mean_absolute_error(y_train, y_pred)
  print("Result for training sample")
  print(f"R2 score: {R2}")
  print(f"MSE: {RMSE}")
  print(f"MAE: {MAE}")
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train, y_train)
```

```python
evaluate_model_train(lr)
evaluate_model(lr)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_estimators=150, random_state=42,max_depth=30)
rf.fit(X_train, y_train)
```

```python
evaluate_model_train(rf)
evaluate_model(rf)
```

```python
from xgboost import XGBRegressor

xgb_model = XGBRegressor()
xgb_model.fit(X_train, y_train)
```

```python
evaluate_model_train(xgb_model)
evaluate_model(xgb_model)
```

```python
y_pred_plot= xgb_model.predict(X_test)
plt.figure(figsize=(15, 8))

plt.subplot(1,2,1)
plt.plot(y_test.values,color="green")
plt.xlabel("Index")
plt.ylabel("Ammonia conversion")
plt.title("Actual")

plt.subplot(1,2,2)
plt.plot(y_pred_plot,color="red")
plt.xlabel("Index")
plt.ylabel("Ammonia conversion")
plt.title("Predicted")
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model=Sequential()
model.add(layers.Dense(256, activation='relu', input_shape=(10,)))
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dense(1,activation="linear"))
```

```python
callback=EarlyStopping(monitor='val_loss', patience=25,verbose=1,restore_best_weights=True)
```

```python
model.compile(loss='mean_squared_error', optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), metrics=['mse'])
```

```python
model.summary()
```

```python
history=model.fit(X_train, y_train, epochs=1000, batch_size=32, validation_data=(X_test, y_test),callbacks=[callback])
```

```python
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.title("Model loss")
plt.ylabel("Loss")
plt.xlabel("Epoch")
plt.legend(["Train", "Validation"], loc="upper left")
```

```python
evaluate_model_train(model)
evaluate_model(model)
```
